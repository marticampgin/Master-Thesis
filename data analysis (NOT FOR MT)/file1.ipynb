```python
europe_sales = pd.read_csv("Europe Sales Records.csv")
```

```python
europe_sales.shape
```

```python
europe_sales.head(10)
```

```python
europe_sales.dtypes
```

```python
# Type category is ideal for columns with a limited number of unique values
# We have 1330+ records, so setting country, item type, and item priority to cetegory might be beneficial
europe_sales.nunique()
```

```python
europe_sales.info()
```

```python
europe_sales['Item Type'] = europe_sales['Item Type'].astype('category')
europe_sales['Country'] = europe_sales['Country'].astype('category')
europe_sales['Order Priority'] = europe_sales['Order Priority'].astype('category')   
```

```python
# Size reduced!
europe_sales.info()
```

```python
europe_sales.dtypes
```

```python
europe_sales.sort_values(["Country", 'Total Profit'], inplace=True, ascending=[True, False])
```

```python
europe_sales.head(10)
```

```python
europe_sales['Profit Rank'] = europe_sales['Total Profit'].rank(ascending=False).astype(int)
```

```python
europe_sales.sort_values("Total Profit", ascending=False).head(10)
```

```python
# This can be actually done upon loading the file by passing date columns in parse_dates as well date_format
europe_sales['Order Date'] = pd.to_datetime(europe_sales['Order Date'], format="%m/%d/%Y")
europe_sales['Ship Date'] = pd.to_datetime(europe_sales['Ship Date'],  format="%m/%d/%Y")
```

```python
europe_sales.head(5)
```

```python
# Filtering data based on conditions
europe_sales[europe_sales['Country'] == 'Latvia'].head(5)
```

```python
estonia_filter = europe_sales['Country'] == 'Estonia'
profit_rank_a50 = europe_sales['Profit Rank'] <= 200

europe_sales[estonia_filter & profit_rank_a50].head(5)
```

```python
target_entities = europe_sales['Item Type'].isin(['Beverages', 'Personal Care'])
europe_sales[target_entities]
```

```python
price_range = europe_sales['Unit Price'].between(80, 500)
europe_sales[price_range]
```

```python
date_range = europe_sales['Order Date'].between("2017-01-01", "2019-01-01")
europe_sales[date_range]
```
